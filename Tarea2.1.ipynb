```python
import pandas as pd
df=pd.read_csv('./altura_peso.csv')
df
```

```python
X=df["altura"]
y=df["peso"]

y
```

```python
X_procesada1 = X.values.reshape(-1,1)
y_procesada1 = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
modelo=LinearRegression()

modelo.fit(X_procesada1,y_procesada1)

```

```python
altura = 175
predicion = modelo.predict([[altura]])
print(f'El peso de {altura} es {predicion}')
```

```python
altura = 168
predicion = modelo.predict([[altura]])
print(f'El peso de {altura} es {predicion}')
```

```python
#¿Funciona bien o no? ¿Por qué cree que es así?
#Funciona bien pero no perfectamente, por que a diferencia de el ejemplo de las temperaturas donde esta era directamente proporcional un parametro de otro, en cambio el peso no depende direcmente de la altura, si no que hay otros factores que tambien infieren en dicho peso
```

```python
modelo.score(X_procesada1,y_procesada1)
```
